```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

#### Data Loading and cleaning

```python
# 1.load data
menu=pd.read_csv('menu.csv')
```

```python
# 2.checking missing value
menu.isnull().sum()
```

```python
# 3.basic info
menu.info()
```

### Descriptive Statistics

### Category Trend Analysis 

```python
# 1. Average calorie content across each category
menu.groupby('Category')['Calories'].mean().sort_values(ascending=False)
```

```python
# 2. Which category has the highest average fat content?
menu.groupby('Category')['Total Fat'].max().sort_values(ascending=False)
```

```python
# 3. What is the average protein content in each category?
menu.groupby('Category')['Protein'].mean().sort_values(ascending=False)
```

```python
# 4. Which category has the most consistent calorie values (lowest standard deviation)?
menu.groupby('Category')['Calories'].std().sort_values()
```

```python
# 5. Which category contains the most high-protein (>30g) items?
menu[menu['Protein']>30].groupby('Category')['Item'].count().sort_values(ascending=False)
```

```python
# 6. Show the count of items in each category.
menu['Category'].value_counts()
```

```python
# 7.  Which category offers the most balanced nutrition (high protein, low fat, moderate calories)?
balanced_items=menu[(menu['Protein']>15) & (menu['Total Fat']<15) & (menu['Calories'].between(250,500))]
balanced_summary=balanced_items['Category'].value_counts()
print("balanced summary=",balanced_summary)
```

### Customer Focused Questions

```python
# 1. Which items has highest calories ?
menu[['Item','Calories']].sort_values(by='Calories',ascending=False).head(5)
```

```python
# 2.What are the top 10 items with the highest protein content?
menu[['Item','Protein']].sort_values(by='Protein',ascending=False).head(10)
```

```python
# 3.Which items contain zero fat but have high sugar?
menu[(menu['Total Fat'] == 0) & (menu['Sugars'] > 30)]
```

```python
# 4.Which food category has the most calorie-dense items?
menu.groupby('Category')['Calories'].mean().sort_values(ascending=False)
```

```python
# 5.  What is the calorie range for each category?
cal_range = menu.groupby('Category')['Calories'].agg(['min', 'max'])
cal_range['range'] = cal_range['max'] - cal_range['min']
print(cal_range.sort_values(by='range', ascending=False))
```

### CUSTOMER BEHAVIOUR (INFERRED FROM PRODUCT) 

```python
# 1. Which types of items are likely to appeal to health-conscious customers?
menu[(menu['Calories'] < 300) & (menu['Protein'] >= 15)][['Item', 'Calories', 'Protein']]
```

#### Egg White Delight has the least calories and more protein,hence health conscious customers can choose this

```python
# Which categories might be preferred by customers looking for quick energy (high sugar)?
menu.groupby('Category')['Sugars'].mean().sort_values(ascending=False)
```

#### The customers who are looking for quick energy(high sugar) can use Smoothies & Shakes

```python
# Which items are best suited for protein-focused diets?
menu[menu['Protein']>30][['Item','Protein','Calories']].sort_values(by='Protein',ascending=False)
```

#### The item which is focused on protein is Chicken McNuggets

```python
# Are there outliers in any category (e.g., one item much higher in calories)?
menu.loc[menu.groupby('Category')['Calories'].idxmax()][['Item','Category','Calories']].sort_values(by='Calories',ascending=False)
```

### Visual Analysis 

```python
# Plot a bar chart showing the average calories per food category. 
menu.groupby('Category')['Calories'].mean().plot(kind='bar',title='Average calories per food category',color='blue')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.show()
```

```python
# Create a pie chart of the total number of items per category.
menu['Category'].value_counts().plot(kind='pie',autopct='%1.f%%',title='Number of items per category')
plt.show()
```

```python
# Compare fat and sugar content using a grouped bar chart for top 5 categories.
top5 = menu['Category'].value_counts().head(5).index
grouped = menu[menu['Category'].isin(top5)].groupby('Category')[['Total Fat', 'Sugars']].mean()
grouped.plot(kind='bar', figsize=(10, 6))
plt.title("Avg Fat & Sugar by Category (Top 5)")
plt.ylabel("Grams")
plt.xticks(rotation=45)
plt.show()
```
